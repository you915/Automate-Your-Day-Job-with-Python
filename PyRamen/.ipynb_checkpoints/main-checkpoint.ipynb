```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path("../Resources/menu_data.csv")
sales_filepath = Path("../Resources/sales_data.csv")

# @TODO: Initialize list objects to hold our menu and sales data
menu = []


# @TODO: Read in the menu data into the menu list

with open(menu_filepath, 'r') as menulist:
    
    reader = csv.reader(menulist)
    next (menulist)
    
    for menu_data in reader:
        
        menu.append(menu_data)
              
sales = []
# @TODO: Read in the sales data into the sales list
with open(sales_filepath, 'r') as saleslist:
    
    reader = csv.reader(saleslist)
    next (saleslist)

    for sales_data in reader:
        
        sales.append(sales_data)
    
# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

menu_counts = dict()
for index in range(len(sales)):
    if sales[index][4] in menu_counts:
        menu_counts[sales[index][4]] += int(sales[index][3])
    else:
        menu_counts[sales[index][4]] = 1

make_dict = dict()
for index in range(len(menu)):
    if menu[index][0] in make_dict:
        make_dict[menu[index][0]] = float(menu[index][3])
    else:
        make_dict[menu[index][0]] = float(menu[index][3])

cost_dict = dict()
for index in range(len(menu)):
    if menu[index][0] in cost_dict:
        cost_dict[menu[index][0]] = float(menu[index][4])
    else:
        cost_dict[menu[index][0]] = float(menu[index][4])

sales_item = menu_counts.keys()

list = []
for i in sales_item:
    list.append(i)

with open('PyRamen.txt', 'w') as f:
    
    for i in range (len(list)):

        report[list[i]] = {}
        if list[i] in menu_counts:
            quantity = menu_counts[list[i]]
            price = make_dict[list[i]]
            cost = cost_dict[list[i]]
        revenue = quantity * price
        cogs = cost * quantity
        profit = revenue - cogs    
        
        report[list[i]]["01-count"] = quantity
        report[list[i]]["02-revenue"] = revenue
        report[list[i]]["03-cogs"] = cogs
        report[list[i]]["04-profit"] = profit
        
        print(list[i], report[list[i]], file=f)

    
```
