```python
# Import csv
import csv

# Import from the pathlib library, the main class Path
from pathlib import Path

# Set the input file path irregardless of OS
csvpath = Path("../Resources/budget_data.csv")

# Create a list to store dates
date_list = []

# Create a list to store profits/losses
profit_losses_list = []

# Initialize all variables
total = 0
average = 0
greatest_increase_in_profits = 0
greatest_decrease_in_profits = 0
profits = 0

# Open the file in "read" mode ('r') and store the
# contents in the variable "text"
with open(csvpath, 'r') as csvfile:
    
    reader = csv.reader(csvfile)
    
    # Add date and profitx/losses information in the list
    for date, profit_losses in reader:
        
        date_list.append(date)
        profit_losses_list.append(profit_losses)
    total_month = len(date_list)-1
    
    # For loop to calculate profit/losses
    for profit_losses in profit_losses_list[1:]:
        total += int(profit_losses)
    
    index = 1
    for profit in range (len(date_list)-2):
        profits_losses = int(profit_losses_list[index+1]) - int(profit_losses_list[index])
        if greatest_increase_in_profits < profits_losses:
            greatest_increase_in_profits = profits_losses
            # Index for greatest increase in profits
            record_index_i = index+1
            index += 1
        elif greatest_decrease_in_profits > profits_losses:
            greatest_decrease_in_profits = profits_losses
            # Index for greatest decrease in profits
            record_index_d = index+1
        else:
            index += 1
    
    # To calculate average change
    average_change = (int(profit_losses_list[len(date_list)-1]) - int (profit_losses_list[1])) / (len(date_list)-2)
    
    # Greatest_increase_in_profits
    greatest_increase_in_profits = int(profit_losses_list[record_index_i]) - int(profit_losses_list[record_index_i-1])
    greatest_increase_in_profits_date = date_list[record_index_i]
    
    # Greatest_decrease_in_profits
    greatest_decrease_in_profits = int(profit_losses_list[record_index_d]) - int(profit_losses_list[record_index_d-1])
    greatest_decrease_in_profits_date = date_list[record_index_d]

a1 = "Financial Analysis"
a2 = "-----------------------"
a3 = f"Total Months: {total_month}"
a4 = f"Total: ${total}"
a5 = f"Average  Change: ${round(average_change,2)}"
a6 = f"Greatest Increase in Profits: {greatest_increase_in_profits_date} (${greatest_increase_in_profits})"
a7 = f"Greatest Decrease in Profits: {greatest_decrease_in_profits_date} (${greatest_decrease_in_profits})"

print (a1)
print (a2)
print (a3)
print (a4)
print (a5)
print (a6)
print (a7)

output_path = Path('PyBank.txt')
with open (output_path, 'w') as saveFile:
    
    a1 = "Financial Analysis\n"
    a2 = "-----------------------\n"
    a3 = f"Total Months: {total_month}\n"
    a4 = f"Total: ${total}\n"
    a5 = f"Average  Change: ${round(average_change,2)}\n"
    a6 = f"Greatest Increase in Profits: {greatest_increase_in_profits_date} (${greatest_increase_in_profits})\n"
    a7 = f"Greatest Decrease in Profits: {greatest_decrease_in_profits_date} (${greatest_decrease_in_profits})\n"

    saveFile.write(a1)
    saveFile.write(a2)
    saveFile.write(a3)
    saveFile.write(a4)
    saveFile.write(a5)
    saveFile.write(a6)
    saveFile.write(a7)
    saveFile.close()
```
